## **Clean the data**

```
#Rename the columns and making them similar
#
columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_NUMBER_INT', 
           'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID']
day1.columns = columns
day1
```

```
# Telcomdataset 2
day2.columns = columns
day2
```

```
# Telcomdataset 3
day3.columns = columns
day3
```

### **Concatenating the three datasets**

```
#Join the three datasets
df_joined = pd.concat([day1, day2, day3], ignore_index = True)
df_joined.head()
```

```
# Drop the columns that are not useful
df_joined.drop(columns= ['COUNTRY_A', 'COUNTRY_B', 'CELL_ON_SITE'], inplace=True)
df_joined.head()
```

#### **Splitting the date_time column to date and time**

```
#Splitting the date_time column and creating two new columns to hold the two variables 
df_joined[['date', 'time']] = df_joined.DATE_TIME.str.split(expand=True)
df_joined.head()
```

```
#Dropping DATE_TIME column
df_joined.drop(columns=['DATE_TIME'], inplace=True)
df_joined.head()
```

#### **Inspecting data in concatenated dataset**

```
# Data in the joined dataset
# groupby product
df_data = df_joined.groupby(['time', 'PRODUCT'])[['PRODUCT']].count()
df_data.head(10)
```

```
df_data.tail(10)
```

```
# Data in the joined dataset
# groupby value
df_data2 = df_joined.groupby(['time', 'PRODUCT'])[['VALUE']].sum()
df_data2.head(10)
```

```
df_data2.tail(10)
```

### **Merge the concatenated dataframe with cells_geo**

```
# Create new with merged dataframes from the concantenated above
#
merged = cells_geo.merge(df_joined, how='left', on='SITE_ID')
merged.head()
```

```
# values of the merged table where time is not null
#
fmerged = merged[merged.time.notnull()]
fmerged.head()
```

```
# Groupby the merged table with no null time values by time and product 

fmerged.groupby(['time', 'PRODUCT'])[['VALUE']].sum().head(10)
```
